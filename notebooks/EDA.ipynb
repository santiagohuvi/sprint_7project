```python
import pandas as pd
import plotly.express as px 
```

```python
data = pd.read_csv('https://practicum-content.s3.us-west-1.amazonaws.com/new-markets/Data_sprint_4_Refactored/vehicles_us.csv')
```

```python
data.head()
```

```python
fig = px.scatter(data, y = 'price', x = 'model_year', hover_data = 'model', color = 'condition')
fig.update_layout(legend_title_text = 'condition')
fig.show()
```
